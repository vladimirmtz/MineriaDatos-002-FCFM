Un árbol de decisión es una estructura que se utiliza para dividir un conjunto de datos que contiene una gran cantidad de registros en conjuntos más pequeños mediante la aplicación de un conjunto de reglas de decisión. En otras palabras, es una estructura que se utiliza aplicando pasos simples de toma de decisiones, dividiendo grandes cantidades de registros en grupos de registros muy pequeños.

### Métricas de evaluación

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve

y_predictions = decTree_model.predict(X_test)

conf_matrix = confusion_matrix(y_predictions, y_test)


print(f'Accuracy: {accuracy_score(y_predictions, y_test)*100}')
print()
print(f'Confussion matrix: \n{conf_matrix}\n')

sns.heatmap(conf_matrix, annot=True)
```

Es una herramienta de medición que proporciona información sobre la precisión de las predicciones

```python
tn = conf_matrix[0,0]
fp = conf_matrix[0,1]
tp = conf_matrix[1,1]
fn = conf_matrix[1,0]

total = tn + fp + tp + fn
real_positive = tp + fn
real_negative = tn + fp
```

TP - Verdadero Positivo: El modelo predijo correctamente la clase positiva como una clase positiva.

FP - Falso positivo: el modelo predijo la clase negativa como una clase de falso positivo.

FN - Falso Negativo: El modelo predijo la clase positiva como falsa, clase negativa.

TN - Negativo verdadero: el modelo predijo correctamente la clase negativa.


```python
accuracy  = (tp + tn) / total # Accuracy Rate
precision = tp / (tp + fp) # Positive Predictive Value
recall    = tp / (tp + fn) # True Positive Rate
f1score  = 2 * precision * recall / (precision + recall)
specificity = tn / (tn + fp) # True Negative Rate
error_rate = (fp + fn) / total # Missclassification Rate
prevalence = real_positive / total
miss_rate = fn / real_positive # False Negative Rate
fall_out = fp / real_negative # False Positive Rate

print(f'Accuracy    : {accuracy*100}')
print(f'Precision   : {precision*100}')
print(f'Recall      : {recall*100}')
print(f'F1 score    : {f1score*100}')
print(f'Specificity : {specificity*100}')
print(f'Error Rate  : {error_rate*100}')
print(f'Prevalence  : {prevalence*100}')
print(f'Miss Rate   : {miss_rate*100}')
print(f'Fall Out    : {fall_out*100}')
```

Tasa de precisión: una medida de la frecuencia con la que el clasificador predice correctamente.

Precisión: muestra cuántos de los valores que adivinamos como Positivos son realmente Positivos.

Recuerde: es una medida de cuánto predice correctamente el clasificador el verdadero valor positivo. También conocido como sensibilidad, precisión o recuperación. (Sensibilidad, Tasa de aciertos o Recuperación) Debe ser lo más alto posible.

Puntuación F1: El valor de la puntuación F1 muestra la media armónica de los valores de precisión y recuperación. La razón por la que es un promedio armónico en lugar de un promedio simple es que no debemos ignorar los casos extremos. Si hubiera un cálculo promedio simple, el puntaje F1 de un modelo con un valor de precisión de 1 y un valor de recuperación de 0 sería 0.5, y esto nos engañaría.

Especificidad: es una medida de cuánto el clasificador predijo correctamente el verdadero valor negativo.

Tasa de clasificación errónea (tasa de error): es una medida de la frecuencia con la que el clasificador adivina incorrectamente. También conocido como tasa de error.

Prevalencia: es la medida de la frecuencia con la que se encuentra un valor de 1 al final de la estimación.

Tasa de fallos: Es la proporción de los que se predice que serán 0 a pesar de que el valor real sea 1. También se conoce como tasa de pérdidas.

Fall out: Es la proporción de los que se predice que será 1 aunque el valor real sea 0.

### AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='orange', label='ROC' )
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
```

Es uno de los criterios de evaluación más importantes para comprobar el rendimiento de cualquier modelo de clasificación. Es una de las métricas más utilizadas para evaluar el rendimiento de los algoritmos de aprendizaje automático, especialmente en los casos en los que existen conjuntos de datos inestables.

```python
probs = decTree_model.predict_proba(X_test)
probs = probs[:, 1]
```

```python
auc = roc_auc_score(y_test, probs)
print('AUC: ', auc*100)
```

Una de las métricas más utilizadas es la curva AUC-ROC. AUC significa "Área bajo la curva ROC". El alcance de esta área es AUC. Cuanto mayor sea el área cubierta, mejores serán los modelos de aprendizaje automático para discriminar las clases impartidas

```python
fpr, tpr, thresholds = roc_curve(y_test, probs)
plt.legend(loc = 'lower right')
plot_roc_curve(fpr, tpr)
```
